```python
import pandas as pd
import numpy as np
from alpha_vantage.timeseries import TimeSeries
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
import warnings
warnings.filterwarnings('ignore')

ts = TimeSeries(key = '7OF1QZ0UX2UELHFW', output_format = 'pandas')
data, meta_data = ts.get_intraday(symbol = 'AMZN', interval = '1min', outputsize = 'full')
AMZN = data['4. close']
```

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import Request
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
news_tables = {}
url = 'https://finviz.com/quote.ashx?t=AMZN'
request = Request(url = url, headers = {'user-agent': 'my-app/0.0.1'})
response = urlopen(request)
html = BeautifulSoup(response)
news_table = html.find(id = 'news-table')
news_tables['AMZN'] = news_table
```

```python
amzn = news_tables['AMZN']
amzn_tr = amzn.findAll('tr')

for i, table_row in enumerate(amzn_tr):
    a_text = table_row.a.text
    td_text = table_row.td.text
    print(a_text)
    print(td_text)
```

```python
parsed_news = []
for file_name, news_table in news_tables.items():
    for x in news_table.findAll('tr'):
        text = x.a.get_text() 
        date_scrape = x.td.text.split()
        if len(date_scrape) == 1:
            time = date_scrape[0]
        else:
            date = date_scrape[0]
            time = date_scrape[1]
        ticker = file_name.split('_')[0]
        parsed_news.append([ticker, date, time, text])
parsed_news
```

```python
vader = SentimentIntensityAnalyzer()
columns = ['ticker', 'date', 'time', 'headline']
parsed_and_scored_news = pd.DataFrame(parsed_news, columns=columns)
scores = parsed_and_scored_news['headline'].apply(vader.polarity_scores).tolist()
scores_df = pd.DataFrame(scores)
parsed_and_scored_news = parsed_and_scored_news.join(scores_df, rsuffix='_right')
parsed_and_scored_news['date'] = pd.to_datetime(parsed_and_scored_news.date).dt.date
news = parsed_and_scored_news[9:]
news.head()
#news.to_csv('news.csv')
```
